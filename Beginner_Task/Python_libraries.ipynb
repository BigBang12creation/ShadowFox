```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Subject': ['DELD', 'DELD', 'OS', 'OS', 'OOPCG', 'OOPCG', 'DS', 'DS'],
    'Marks':   [35, 50, 60, 70, 70, 60, 80, 60],
    'Gender':  ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    }
df = pd.DataFrame(data)
sns.barplot(x='Subject', y='Marks', hue='Gender', data=df, palette='Set2')
plt.title('Marks Distribution by Gender')
plt.xlabel('Subject name')
plt.ylabel('Marks')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Bill_Length': [35,45,40,21,50,55,63,30,70],
    'Frequency': [2, 3, 1, 4, 5, 2, 7, 5, 3]
}
df = pd.DataFrame(data)
sns.histplot(x='Bill_Length', weights='Frequency',data=df,color='blue')
plt.title('Histogram Plot of Bill Length')
plt.xlabel('Bill Length')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'OS': [70, 80, 60, 90, 85],
    'OOPCG': [75, 85, 65, 95, 88],
    'DS': [68, 78, 72, 88, 80],
    'DELD': [90, 95, 85, 98, 92]
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(6,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Student Data')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

sns.regplot(x="sepal_length", y="petal_length", data=df)
plt.title("Regression Plot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```
